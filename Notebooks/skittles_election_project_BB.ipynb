```python
import district_module
import pandas as pd
from io import StringIO
```

```python
response = district_module.get_url_text('https://www.opensecrets.org/races/summary.csv?cycle=2020&id=TN07')
tn07_df = pd.read_csv(StringIO(str(response)))
tn07_df.to_csv(r'C:\Users\billy\Documents\NSS_projects\open-secrets-skittles\Data\tn07_df.csv', index=False)
```

```python
abbrev_table = district_module.get_wiki_table('https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations')
district_table = district_module.get_wiki_table('https://en.wikipedia.org/wiki/2020_United_States_House_of_Representatives_elections')
```

```python
abbrev_df = (
    district_module
    .convert_wikitable_to_df(abbrev_table, 1)[['Name', 'USPS']]
    .rename(columns={'Name': 'State', 'Unnamed: 5_level_1':'State Initial'})
)
```

```python
district_df =  (
    district_module
    .convert_wikitable_to_df(district_table, 1)[['State', 'Total seats']]
    .rename(columns={'Total seats': 'Districts'})
)
```

```python
all_states_districts = (
    pd
    .merge(abbrev_df, district_df)
    .droplevel(0, axis = 1)
    .drop_duplicates(subset='State')
)
```

```python
districts_per_state = all_states_districts['Districts'].tolist()
state_initials = all_states_districts['State Initial'].tolist()
```

```python
states_districts_df = district_module.get_districts_df('https://www.opensecrets.org/races/summary.csv?cycle=2020&id=', districts_per_state, state_initials)
states_districts_df.to_csv(r'C:\Users\billy\Documents\NSS_projects\open-secrets-skittles\Data\states_districts_df.csv', index=False)
```

```python
tn_df = states_districts_df[states_districts_df['State'] == 'Tennessee']
tn_df.to_csv(r'C:\Users\billy\Documents\NSS_projects\open-secrets-skittles\Data\tn_df.csv', index=False)
```
